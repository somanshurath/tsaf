```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)
```

```python
# white noise time series w/ normal distribution
n1 = 1000
x1 = np.random.randn(n1)
plt.plot(x1)
plt.show()
```

```python
# white noise time series from uniform distribution
n2 = 1000
x2 = np.random.rand(n2)
plt.plot(x2)
plt.show()
```

```python
n3 = 1000
x3 = np.random.exponential(1, n3)
plt.plot(x3)
plt.show()
```

```python
# random walk
n4 = 1000
x4 = np.cumsum(np.random.randn(n4))
plt.plot(x4)
plt.show()
```

```python
# random walk with drift
n5 = 1000
x5 = np.cumsum(np.random.randn(n5) + 0.01)
plt.plot(x5)
plt.show()
```

```python
# random walk with drift and trend
n6 = 1000
x6 = np.cumsum(np.random.randn(n6) + 0.01) + np.linspace(-10, 10, n6)
plt.plot(x6)
plt.show()
```

```python
# random walk with drift and seasonality
n7 = 1000
x7 = np.cumsum(np.random.randn(n7)) + np.sin(np.linspace(0, 10, n7))
plt.plot(x7)
plt.show()
```

```python
# ar(1) process
n8 = 1000
x8 = np.zeros(n8)
for i in range(1, n8):
    x8[i] = 0.9 * x8[i - 1] + np.random.randn()
plt.plot(x8)
plt.show()
```

```python
# ar(2) process
n9 = 1000
x9 = np.zeros(n9)
for i in range(2, n9):
    x9[i] = 0.5 * x9[i - 1] + 0.3 * x9[i - 2] + np.random.randn()
plt.plot(x9)
plt.show()
```

```python
# sinusoidal time series
n10 = 1000
x10 = np.sin(np.linspace(0, 10, n10))
x11 = np.cos(np.linspace(0, 10, n10))
plt.plot(x10)
plt.plot(x11)
plt.show()
```

```python
# find the autocorrelation of the time series
from statsmodels.tsa.stattools import acf
acf_res = acf(x10, nlags=1000)
plt.bar(range(len(acf_res)), acf_res)
plt.show()
# find the partial autocorrelation of the time series
from statsmodels.tsa.stattools import pacf
pacf_res = pacf(x10, nlags=500)
plt.bar(range(len(pacf_res)), pacf_res)
plt.show()
```

```python
# martingale process where E[x(t) | x(t-1), x(t-2), ...] = x(t-1)
n11 = 1000
x11 = np.zeros(n11)
for i in range(1, n11):
    x11[i] = x11[i - 1] + np.random.randn()
plt.plot(x11)
plt.show()
```
