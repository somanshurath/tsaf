```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
np.random.seed(1)
y = np.random.normal(0, 1, 100)
for i in range(1, 100):
    y[i] = y[i-1] + 0.01 + y[i]
x = np.arange(100)
df = pd.DataFrame({'y': y, 'x': x})
```

```python
fig = plt.figure(figsize=(12, 8))
plt.plot(y)
plt.xlabel('Time')
plt.ylabel('y')
plt.title('Random Walk with Drift')
plt.show()
```

```python
X = sm.add_constant(x)
model = sm.OLS(y, X)
results = model.fit()
```

```python
true_mean = np.zeros(100)
true_mean[0] = 0
for i in range(1, 100):
    true_mean[i] = true_mean[i-1] + 0.01
```

```python
plt.figure(figsize=(12, 8))
plt.plot(y)
plt.plot(results.fittedvalues)
plt.plot(true_mean)
plt.xlabel('Time')
plt.ylabel('y')
plt.title('Random Walk with Drift')
plt.show()
```
